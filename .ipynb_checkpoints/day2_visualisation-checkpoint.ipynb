```python
# %matplotlib notebook
%matplotlib inline
def group_barplot(feat_groupby, feat_agg='price_value', agg_funcs=[np.mean, np.median, np.size], feat_sort='mean', top=50, subplots=True):
    return(
        df
        .groupby(feat_groupby)[feat_agg]
        .agg(agg_funcs)
        .sort_values(by=feat_sort, ascending=False)
        .head(top)
    
    ).plot(kind='bar', figsize=(15,6), subplots=subplots)
```

```python
group_barplot('param_marka-pojazdu')
```

```python
group_barplot('param_kraj-pochodzenia', feat_sort='size')
```

```python
group_barplot('param_kolor')
```
